```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
df_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([df_2020, df_2021], ignore_index=True)

# Reset the index
combined_df.reset_index(drop=True, inplace=True)

combined_df
```

```python
# Check if any values are null.
null_values = combined_df.isnull().sum()
null_values
```

```python
# Check the data type of each column
data_types = combined_df.dtypes
data_types
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'], errors='coerce')

# Confirm the data type has been changed
invoice_date_type = combined_df['invoice_date'].dtype
print(invoice_date_type)
```

```python
# Confirm that the "invoice_date" data type has been changed.
data_types = combined_df.dtypes
data_types
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
# Group by region, state, and city, and sum the number of units sold
grouped_df = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the summed column to "Total_Products_Sold"
grouped_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)


# Show the top 5 results.
top_5_results = grouped_df.nlargest(5, 'Total_Products_Sold')
top_5_results
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.


# Rename the "units_sold" column to "Total_Products_Sold"


# Show the top 5 results.
# Create a pivot table to show the number of products sold for region, state, and city
pivot_table = combined_df.pivot_table(
    index=['region', 'state', 'city'],
    values='units_sold',
    aggfunc='sum'
).reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results
top_5_results = pivot_table.nlargest(5, 'Total_Products_Sold')

top_5_results
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.

total_sales_df = combined_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the "total_sales" column to "Total Sales"
total_sales_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_5_sales_results = total_sales_df.nlargest(5, 'Total Sales')
top_5_sales_results


```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.

pivot_table_sales = combined_df.pivot_table(
    index=['region', 'state', 'city'],
    values='total_sales',
    aggfunc='sum'
).reset_index()

pivot_table_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results
top_5_pivot_results = pivot_table_sales.nlargest(5, 'Total Sales')

top_5_pivot_results
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
pivot_table_invoice = combined_df.pivot_table(
    index='invoice_date',
    values='total_sales',
    aggfunc='sum'
)

# Optionally rename the "total_sales" column to "Total Sales"
pivot_table_invoice.rename(columns={'total_sales': 'Total Sales', 'invoice_date':'Invoice Date'}, inplace=True)
pivot_table_invoice
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.


# Sort the resampled pivot table in descending order on "Total Sales".

daily_sales = pivot_table_invoice.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales"
sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=False)
sorted_daily_sales
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.


# Sort the resampled pivot table in descending order on "Total Sales".

weekly_sales = pivot_table_invoice.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales"
sorted_weekly_sales = weekly_sales.sort_values(by='Total Sales', ascending=False)
sorted_weekly_sales
```
